```python
# Downloading Dataset

import kagglehub

# Download latest version

import os

os.makedirs("data", exist_ok=True)

path = kagglehub.dataset_download("brsahan/vehicle-co2-emissions-dataset")

print("Path to dataset files:", path)
```

```python
path
```

```python
import pandas as pd
import numpy as np

emissions = pd.read_csv(path + "/co2.csv")
emissions.head()
```

```python
emissions.info()
```

```python
emissions.describe().T
```

```python
emissions.nunique()
```

```python
veh_classes = emissions["Vehicle Class"].unique()
veh_classes
```

```python
# Simplifying the vehicle class names by removing spaces and hyphens, and converting to lowercase for easier handling in code.
veh_classes_rename = {
    veh_class : veh_class.replace(" ", "").replace("-","_").lower() for veh_class in veh_classes
}
veh_classes_rename
```

```python
emissions["Vehicle Class"] = emissions["Vehicle Class"].replace(veh_classes_rename)
```

```python
emissions.info()
```

```python
emissions["Make"].unique()
```

```python
emissions["Make"] = emissions["Make"].str.lower().replace(r"[^a-zA-Z0-9]", "_", regex=True)
```

```python
emissions["Make"].unique()
```

```python
emissions
```

```python
emissions.drop(columns=["Model"], inplace=True) # does not matter to our analysis
# emissions.drop(columns=["Transmission"], inplace=True) # Yet to review if this is useful for analysis
```

```python
emissions
```

```python
emissions.rename(
    columns= {
        "Vehicle Class": "Vehicle_type",
        "Engine Size(L)": "Engine_size_L",
        "Fuel Type": "Fuel_type",
        "Fuel Consumption City (L/100 km)": "Fuel_consumption_city_L_per_100km",
        "Fuel Consumption Hwy (L/100 km)": "Fuel_consumption_hwy_L_per_100km",
        "Fuel Consumption Comb (L/100 km)": "Fuel_consumption_comb_L_per_100km",
        "Fuel Consumption Comb (mpg)": "Fuel_consumption_comb_mpg",
        "CO2 Emissions(g/km)": "CO2_emissions_g_per_km"
    },
    inplace=True
)
```

```python
emissions
```

```python
emissions.info()
```

```python
emissions["Fuel_type"].unique()
```

```python
emissions["Fuel_type"] = emissions["Fuel_type"].replace(
    {
        "X": "gasoline",
        "Z": "electricity",
        "D": "diesel",
        "E": "ethanol",
        "N": "natural gas"
    }
)
```
